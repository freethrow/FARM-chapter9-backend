```python
import pandas as pd
```

```python
cars = pd.read_csv('filteredCars.csv')
```

```python
cars.head()
```

```python
cars.describe()
```

```python
cars['brand'] = cars['brand'].astype('str')
cars['make'] = cars['make'].astype('str')
```

```python
cars.dtypes
```

```python
# brand dummies
brand_dummies = pd.get_dummies(cars['brand'])
```

```python
brand_dummies.head()
```

```python
make_dummies = pd.get_dummies(cars['make'])
```

```python
make_dummies.head()
```

```python
# concatenate axis 1
processed = pd.concat([cars, brand_dummies, make_dummies], axis=1)
```

```python
processed.head()
```

```python
data = processed.drop(['brand', 'make'], axis=1)
```
